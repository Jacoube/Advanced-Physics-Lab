```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import odr
import pandas as pd
from pathlib import Path
from scipy.optimize import fsolve
```

```python
# Set up paths from directory
dir_path = Path.cwd()
data_dir = dir_path / "Data"

df = pd.read_csv(data_dir / "IntroResonanceCurveCSV.csv")
capacitance = df["Capacitance (nF)"]
amp = df["V_R (mV) Amplitude"] / 2
peak_to_peak = df["V_R (mV) Peak-to-Peak"] / 2
```

```python
# Estimate error as peak-to-peak - amp / 2
amp_err = (peak_to_peak - amp)
```

```python
# Plot
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

plt.errorbar(capacitance, amp, 
    # xerr=resistance_uncerts[1:], 
    yerr=amp_err,
    fmt='o', label='$V_R$', capsize=3, markersize=4)


ax.set_title("Resonance Curve for a Parallel RLC Circuit")

ax.legend()

ax.set_xlabel(r'Capactiance (nF)')
ax.set_ylabel(r'Amplitude $V_R$')
ax.grid(True, linestyle='--', alpha=0.7)

# Display the plot
fig.tight_layout()
```
